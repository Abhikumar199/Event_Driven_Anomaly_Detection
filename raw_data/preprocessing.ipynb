### Smoothing the dataset

```python
window = 5

df1['TDS_smooth'] = df1['TDS_raw'].rolling(window).mean()
df1['Turbidity_smooth'] = df1['Turbidity_raw'].rolling(window).mean()


df1 = df1.dropna()
```

```python
Q1 = df1['Turbidity_raw'].quantile(0.25)
Q3 = df1['Turbidity_raw'].quantile(0.75)
IQR = Q3 - Q1

df1 = df1[(df1['Turbidity_raw'] >= Q1 - 1.5*IQR) &
        (df1['Turbidity_raw'] <= Q3 + 1.5*IQR)]
```

```python
df1
```

```python
df1=df1.drop(["TDS_raw","Turbidity_raw"],axis=1)
```
